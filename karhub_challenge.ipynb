```python
# Cleaning the data
```

```python
from functools import reduce
from toolz import pipe
import re
from unidecode import unidecode

def camel_to_snake_case(string: str) -> str:
    reduced = reduce(
        lambda char, next_char: char + next_char + ("_" if (not char.isupper()) and next_char.isupper() else ""), string[::-1],
    ).lower()[::-1]
    
    if reduced.endswith("_"):
        reduced = reduced[:-1]
    if reduced.startswith("_"):
        reduced = reduced[1:]
    
    return reduced

def sanitize_string(string: str) -> str:
    return unidecode(
        re.sub(
            "[\s+]",
            "",
            string
                .replace("[^a-zA-Z0-9]+", "")
                .replace("(", "")
                .replace(")", "")
        )
    )
```

```python
for column in df_full.columns:
    new_column = pipe(
        column,
        camel_to_snake_case,
        sanitize_string
    )
    df_full.rename(columns={column: new_column}, inplace=True)
```

```python
import json

with open("service_account.json") as file:
    service_account = json.load(file)
```

```python
import pandas_gbq as pbq
from google.oauth2.service_account import Credentials
```

```python
pbq.to_gbq(
    df_full,
    "cadastro_produto.kh_data_engineer_teste_pedroNobrega",
    project_id=service_account["project_id"],
    credentials=Credentials.from_service_account_info(service_account),
    if_exists="replace"
)
```

```python
from google.cloud.bigquery import Client

bq_client = Client(
    project=service_account["project_id"], 
    credentials=Credentials.from_service_account_info(service_account)
)

unique_products = bq_client.query("""
    SELECT DISTINCT 
        nomesku
        , codigo
    FROM `karhub-data-engineer-test`.`cadastro_produto`.`kh_data_engineer_teste_pedroNobrega`
""").result().to_dataframe()

print(unique_products)

unique_vehicles =  bq_client.query("""
    SELECT DISTINCT
        nomesku
        , codigo
        , marca
        , modelo
        , count(1) as qtd_veiculos
    FROM `karhub-data-engineer-test`.`cadastro_produto`.`kh_data_engineer_teste_pedroNobrega`
    GROUP BY
        nomesku
        , codigo
        , marca
        , modelo
    ORDER BY
        qtd_veiculos DESC
""").result().to_dataframe()

print(unique_vehicles)

unique_products_by_category = bq_client.query("""
    SELECT DISTINCT
        nomesku
        , codigo
        , categoria
        , count(1) as qtd_produtos
    FROM `karhub-data-engineer-test`.`cadastro_produto`.`kh_data_engineer_teste_pedroNobrega`
    GROUP BY
        nomesku
        , codigo
        , categoria
    ORDER BY
        qtd_produtos DESC
""").result().to_dataframe()

print(unique_products_by_category)
```
